```python
import pandas as pd
from sklearn.metrics import mean_absolute_error

from sklearn.metrics import roc_auc_score, precision_recall_curve, auc
```

```python
for i in range(3):
  df1 = pd.read_csv(f"../intermediate_submissions/genetherapy_dataset{i}_1.csv")
  df2 = pd.read_csv(f"../intermediate_submissions/genetherapy_dataset{i}_2.csv")

  merged = pd.merge(
    df1[["transcript_id", "transcript_position", "score"]],
    df2[["transcript_id", "transcript_position", "score"]],
    on=["transcript_id", "transcript_position"],
    suffixes=("_model1", "_model2")
)
  
  merged["abs_error"] = (merged["score_model1"] - merged["score_model2"]).abs()
  
  mae = mean_absolute_error(merged["score_model1"], merged["score_model2"])
  min_error = merged["abs_error"].min()
  max_error = merged["abs_error"].max()

  print(f"Dataset {i}")
  print(f"Mean Absolute Error between models: {mae}")
  print(f"Min  Absolute Error: {min_error}")
  print(f"Max  Absolute Error: {max_error}")
```

```python
dataset0_1 = pd.read_csv(f"../intermediate_submissions/genetherapy_dataset0_2.csv")

labels = pd.read_csv(f"../data/data.info.labelled")

dataset0 = pd.merge(dataset0_1, labels, on=['transcript_id', 'transcript_position'], how='left')

y = dataset0['label']
y_pred = dataset0['score']

roc_auc = roc_auc_score(y, y_pred)
precision, recall, _ = precision_recall_curve(y, y_pred)
pr_auc = auc(recall, precision)

print("Dataset 0 evaluation")
print(f"ROC-AUC: {roc_auc:.4f}")
print(f"PR-AUC : {pr_auc:.4f}")
```
